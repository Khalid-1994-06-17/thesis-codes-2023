```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t1=np.linspace(0,4*np.pi,200)
t2=np.linspace(4*np.pi,5*np.pi,200)
t3=np.linspace(5*np.pi,9*np.pi,200)

t1a=np.linspace(0,2*np.pi,200)
t2a=np.linspace(2*np.pi,4*np.pi,20)
t3a=np.linspace(4*np.pi,6*np.pi,200)

x1=np.sin(t1)
x2=0*np.ones(200)
x3=np.sin(t3)

x4=np.sin(7*t2)

x1a=np.sin(4*t1a)+2.05
x2a=np.sin(4*t2a)-0.5*t2a+5.2
x3a=np.sin(4*t3a)-1.1

plt.figure(figsize=(30, 5))
plt.title("Shaplet Outlier")
plt.plot(t1,x1, color='blue')
plt.plot(t2,x2, color='red')
plt.plot(t3,x3, color='blue')
plt.grid()
plt.savefig("Shaplet Outlier.png")

plt.figure(figsize=(30, 5))
plt.title("Seasonal Outlier")
plt.plot(t1,x1, color='blue')
plt.plot(t2,x4, color='red')
plt.plot(t3,x3, color='blue')
plt.grid()
plt.savefig("Seasonal Outlier.png")

plt.figure(figsize=(30, 5))
plt.title("Trend Outlier")
plt.plot(t1a,x1a, color='blue')
plt.plot(t2a,x2a, color='red')
plt.plot(t3a,x3a, color='blue')
plt.grid()
plt.savefig("Trend Outlier.png")
```
